**Image to Pencil Sketch with Python**

**Importing all the necessary required libraries**

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
image = cv2.imread("gdrive/My Drive/Colab Notebooks/Lets Grow More Internship Tasks/Dataset Task Files/photo taken.jpeg")
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(4,4))
plt.imshow(image)
plt.axis("off")
plt.title("The Original Image")
plt.show()
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(4,4))
plt.imshow(image_gray,cmap="gray")
plt.axis("off")
plt.title("The GrayScale Image")
plt.show()
```

```python
image_invert = cv2.bitwise_not(image_gray)
plt.figure(figsize=(4,4))
plt.imshow(image_invert,cmap="gray")
plt.axis("off")
plt.title("The Inverted Image")
plt.show()
```

```python
image_smoothing = cv2.GaussianBlur(image_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(4,4))
plt.imshow(image_smoothing,cmap="gray")
plt.axis("off")
plt.title("The Smoothen Image")
plt.show()
```

```python
final = cv2.divide(image_gray, 255 - image_smoothing, scale=255)
plt.figure(figsize=(4,4))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("The Final Sketch Image")
plt.show()
```
