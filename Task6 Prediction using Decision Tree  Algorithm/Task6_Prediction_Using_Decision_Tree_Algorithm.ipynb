**Importing all the necessary required libraries**

```python
import sklearn.datasets as datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, plot_confusion_matrix
from sklearn import tree
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**Performing pair plotting to view relationship between the features present in the data**

```python
sns.pairplot(df, hue='Species')
```

**Splitting the data into train and test set**

```python
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
y = df['Species'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, random_state=1)
```

**Training the decision tree classifier**

```python
#Scikit learn decision tree model training
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)  
```

**Testing the model trained on test set**

```python
prediction = clf.predict(X_test)
prediction
```

**Performing the Evaluation on Test set**

```python
#Evaluation for multi class classification
print(classification_report(y_test, prediction))
```

**Plotting the Confusion Matrix**

```python
plot_confusion_matrix(clf, X_test, y_test)
```

**Importing the necessary libraries for graph viz**

```python
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

**Visualizing the graph**

```python
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

**Feeding new data to the classifier to predict the right class**
